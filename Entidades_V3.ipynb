```python
import numpy as np
import pandas as pd
import spacy
from spacy.tokens import Doc, Span

archivo = pd.read_excel("listado de especificaciones para proyecto.xlsx", sheet_name= "Especificaciones sin código ine")
medicamentos = archivo
medicamentos.head(3)
```

```python
nombre_medicamentos = medicamentos["NOMBRE"].tolist()
```

```python
import unicodedata
import re
regex = r'^([A-Za-zÀ-ÿ\s]+)(?:\s+[A-Za-zÀ-ÿ\s]+)*\s*'


def limpiar_texto(texto):
    texto = unicodedata.normalize('NFD', texto)  # Elimina tildes
    texto = ''.join([c for c in texto if not unicodedata.combining(c)])  # Elimina tildes
    texto = texto.lower()  # Convierte a minúsculas
    return texto.strip() 

# Aplicar la expresión regular y limpiar el texto a cada elemento de la lista
nombres_medicamentos_reg = [limpiar_texto(re.match(regex, medicamento).group(1)) for medicamento in nombre_medicamentos]

# Imprimir los nombres de los medicamentos
for nombre in nombres_medicamentos_reg:
    print(nombre)
```

```python
import spacy
from spacy.matcher import Matcher
from spacy.tokens import Span

# Cargar un modelo de spaCy
nlp = spacy.load("es_core_news_lg")

# Obtener el componente ner del pipeline
ner = nlp.get_pipe("ner")

# Crear un diccionario de patrones de medicamentos

# Crear un Matcher
matcher = Matcher(nlp.vocab)

# Agregar los patrones de medicamentos al Matcher
for medicamento, patron in medicamento_patterns.items():
    matcher.add(medicamento, [patron])

#for presentacion, patron_p in presentaciones_patterns.items():
#    matcher.add(presentacion, [patron_p])

```

```python
# Definir una función para manejar las coincidencias del Matcher
def on_match_callback(matcher, doc, i, matches):
    match_id, start, end = matches[i]
    span = Span(doc, start, end, label="MEDICAMENTO")
    doc.ents = [e for e in doc.ents if not (e.start < span.end and e.end > span.start)]  # Eliminar entidades superpuestas
    doc.ents = list(doc.ents) + [span]  # Agregar la nueva entidad

# Procesar un texto
text = "COMPRA DE 1000 LORAZEPAM 2MG/ML AMPOLLA 2ML"
doc = nlp(text)

matches = matcher(doc)
for match_id, start, end in matches:
    on_match_callback(matcher, doc, 0, [(match_id, start, end)])
    
# Imprimir las entidades encontradas en el documento
for ent in doc.ents:
    print("Nombre:", ent.text,"--->","Entidad :" ,ent.label_)


```

```python
id_token = 1

# Iterar a través de los tokens en el documento
for token in doc:
    entity_info = f"{token.ent_iob_}-{token.ent_type_}" if token.ent_iob_ != 'O' else 'O'
    print(f"ID: {id_token} - Token: {token.text} - Entidad: {entity_info}")
    id_token += 1
```
